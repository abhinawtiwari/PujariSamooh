```python
def calc_target (row):
    if (row['vehicle'] == 1) & (row['speed'] < 40) & ((row['pressure'] >= 116) & (row['pressure'] <= 131)) & (row['fuel'] == 1 or row['fuel'] == 2) & ((row['engine-temperature'] >= 195) & (row['engine-temperature'] <= 220)):
        return 0
    elif (row['vehicle'] == 2) & (row['speed'] < 40) & ((row['pressure'] >= 116) & (row['pressure'] <= 131)) & ((row['fuel'] == 1) or (row['fuel'] == 2)) & ((row['engine-temperature'] >= 195) & (row['engine-temperature'] <= 220)):
        return 0
    elif (row['vehicle'] == 3) & (row['speed'] < 40) & ((row['pressure'] >= 25) & (row['pressure'] <= 35)) & ((row['fuel'] == 1) or (row['fuel'] == 2)) & ((row['engine-temperature'] >= 155) & (row['engine-temperature'] <= 220)):
        return 0
    else:
        return 1
```

```python
df['inspect'] = df.apply (lambda row: calc_target(row), axis=1)
df.head()
```
